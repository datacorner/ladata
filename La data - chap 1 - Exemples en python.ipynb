# Loi normale standard

```python
from scipy.integrate import quad
import matplotlib.pyplot as plt
import scipy.stats
import numpy as np

plt.figure(figsize=(20,10))

x_min, x_max = -4.0, 4.0
mean = 0.0 
std = 1.0

x = np.linspace(x_min, x_max, 100)
y = scipy.stats.norm.pdf(x, mean, std)
plt.plot(x,y, color='red')

plt.xlim(x_min,x_max)
plt.ylim(0, 0.45)
plt.title('Loi normale standard',fontsize=10)
plt.show()
```

```python
from scipy.integrate import quad
import matplotlib.pyplot as plt
import scipy.stats
import numpy as np

plt.figure(figsize=(20,10))

def TraceLoiNormale(std, mean, color):
  x_min, x_max = -4.0, 4.0
  mean = mean 
  std = std

  x = np.linspace(x_min, x_max, 100)
  y = scipy.stats.norm.pdf(x, mean, std)
  plt.plot(x,y, color=color)

  plt.xlim(x_min,x_max)
  plt.ylim(0, 0.45)
  plt.title('Loi normale standard',fontsize=10)

TraceLoiNormale(1.0, 0.0, 'red')
TraceLoiNormale(2.0, 0.0, 'blue')
TraceLoiNormale(3.0, 0.0, 'green')

plt.show()
```
